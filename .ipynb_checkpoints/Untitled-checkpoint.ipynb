# Data cleaning

```python
df.drop(['fare', 'home.dest', 'name'], axis=1, inplace=True)
```

```python
df['age'] = df['age'].fillna(0)
```

# Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.hist(bins=20, figsize=(20,15))
plt.show()
```

# OUTLIERS

```python
df_copy = df.copy()
```

```python
df_copy['age'].iloc[0:10] = 500
```

```python
df_copy.head(n=15)
```

```python
df_copy.hist(bins=20, figsize=(20,15))
plt.show()
```
